```python
## settings
import numpy as np
import pandas as pd
import os

from tensorflow.keras.models import load_model
import efficientnet.tfkeras as efn
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from PIL import Image
```

```python
## load data
submission_df = pd.read_csv('./data/sample_submission.csv', index_col='index')
test_df = pd.read_csv('./data/test_df.csv', index_col='index')

dir_path = './data/test'
```

```python
IMAGE_SIZE = 256 

pretrained_network = efn.EfficientNetB4(weights='imagenet',
                                        include_top=False,
                                        input_shape=(IMAGE_SIZE,IMAGE_SIZE,3),
                                        classes=88)

class_model = load_model('./model/0510_class_model.h5',
                       custom_objects={"pretrained_network": pretrained_network})
```

```python
train_dir = './data/class_model_full/train'
class_labels = list(ImageDataGenerator().flow_from_directory(train_dir, class_mode='categorical').class_indices.keys())
```

```python
# i=0
class_pred_list = []
for num, f_name in zip(range(len(test_df)) ,test_df['file_name']):
    try:
        image_path = os.path.join(dir_path, f_name)
        image_path = image_path.replace('\\','/')
        image = Image.open(image_path)
        image = image.resize((256,256))
        image = np.array(image)
        x_test = np.array([image])
        x_test = (x_test)/255
        y_predict = class_model.predict(x_test)
        label = class_labels[y_predict[0].argmax()]
        class_pred_list.append(label)
#         print(submission_df.iloc[num]['class'])
#         submission_df.iloc[num]['class'] = label
#         print(submission_df.iloc[num]['class'])
#         i+=1
        
    except:
            try:
                image_path = os.path.join(dir_path, f_name)
                image_path = image_path.replace('\\','/')
                image = Image.open(image_path)
                image = image.resize((256,256))
                image = np.array(image)
                stacked_image = np.stack((image,)*3, axis=-1)
                x_test = np.array([stacked_image])
                x_test = (x_test)/255
                y_predict = class_model.predict(x_test)
                label = class_labels[y_predict[0].argmax()]
                class_pred_list.append(label)
#                 submission_df.iloc[num]['class'] = label
            
            except:
                print(num, f_name)
#     if i > 5 : break

# print(class_pred_list)
# print(pd.DataFrame(class_pred_list))
submission_df['class'] = pd.DataFrame(class_pred_list)
display(submission_df)
```

```python
submission_df['class'] = pd.DataFrame(class_pred_list)
display(submission_df)
```
